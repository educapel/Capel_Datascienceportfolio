```python
def extracting_data(id_inmueble):

    print('\n casa numero: ' + id_inmueble)

    url = "https://www.idealista.com/inmueble/" + id_inmueble + "/"

    browser.get(url)

    html = browser.page_source

    soup = bs(html, 'lxml')

    

    titulo = soup.find("span",{"class": "main-info__title-main"}).text.strip().split(',')[0]

    print('\n Titulo:' + titulo)

    location = soup.find('span', {'class': 'main-info__title-minor'}).text.split(',')[0]

    precio = soup.find('div', {'class': 'info-data'}).find('span', {'class': 'info-data-price'}).text
    precio = int(precio.replace('.', '').replace('€', '').strip())
    
    feature_1 = soup.find('section', {'id': 'details'}).find('div', {'class': 'details-property-feature-one'})
    f_1 = [features.text.strip() for features in feature_1.find_all('li')]
    features_2 = soup.find('section', {'id': 'details'}).find('div', {'class': 'details-property-feature-two'})
    f_2 = [features.text.strip() for features in features_2.find_all('li')]


    houses['title'] = titulo
    houses['location'] = location
    houses['precio'] = precio

    houses['basic_features'] = f_1
    houses['extra_features'] = f_2

    df_houses_07= pd.DataFrame(houses)

    return(df_houses_07.T)
 





```

```python
df_houses_07 = extracting_data(ids_casas_07.iloc[0].id)

for i in range(1, len(ids)):
    df_houses_07= pd.concat([df_houses_07, extracting_data(ids[i])]) 

    time.sleep(random.randint(4,8))
```
